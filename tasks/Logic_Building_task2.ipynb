```python
#Problem1: Unique Words in a Sentence
def find_unique_words(sentence):
    words=sentence.split() 
    unique_words=set(words) 
    return unique_words

sentence=input("Enter sentence: ")
unique_words=find_unique_words(sentence) 
print(f"Unique words count:{len(unique_words)}")
print(f"Unique words:{unique_words}")
```

```python
#Problem2: Highest Salary from Employee Data
def find_highest_salary(employees):
    highest_paid_employee=max(employees,key=employees.get) 
    highest_salary=employees[highest_paid_employee]    
    return highest_paid_employee,highest_salary

employees={
    "Ravi": 75000,
    "Anita": 68000,  
    "Kiran": 72000
}
name,salary=find_highest_salary(employees)  
print(f"Highest Salary:{name}-{salary}")
```

```python
#Problem3: Find Maximum and Minimum Values
def find_max_min(numbers):  
    max_value=numbers[0]
    min_value=numbers[0]
    for num in numbers:     
        if num>max_value:
            max_value=num
        if num<min_value:
            min_value=num
    return max_value,min_value
    
numbers=[45,22,89,10,66]
maximum,minimum=find_max_min(numbers) 
print(f"List:{numbers}")
print(f"Max:{maximum}")
print(f"Min:{minimum}")
```

```python
#Problem4: Count Products Above a Price Threshold
def count_products_above_threshold(prices,threshold=1000):
    count=0 
    for price in prices:       
        if price>threshold:
            count+=1
    return count

prices=[450,1200,899,1500,300]
result=count_products_above_threshold(prices)  
print(f"Products above 1000:{result}")
```
